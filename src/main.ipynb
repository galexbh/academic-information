```python
import pandas as pd

df_student = pd.read_csv("/content/drive/MyDrive/UNAH/IA/student.csv")
df_ratings = pd.read_csv("/content/drive/MyDrive/UNAH/IA/ratings.csv")
df_teacher = pd.read_csv("/content/drive/MyDrive/UNAH/IA/teacher.csv")
df_race = pd.read_csv("/content/drive/MyDrive/UNAH/IA/race.csv")
facts = pd.read_json("/content/drive/MyDrive/UNAH/IA/facts.json")
```

```python
"""
    It is desired to have as a metric the 10 students with the highest academic average
"""
facts_average = facts.merge(df_ratings, how="inner", on="id_ratings")
facts_average["average"] = (facts_average["note_1"] + facts_average["note_2"] + facts_average["note_3"] + facts_average["note_4"]) / 4
facts_student_average = facts_average.sort_values(by=["average"], ascending=False)
facts_student_average.head(10).merge(df_student, how="inner", on="id_student").iloc[:,[7,9,10]]
```

```python
"""
    Of the teachers, we want to know the full name, address and the first 5 with the lowest salary.
"""

facts_dropped_teacher = facts.merge(df_teacher, how="inner", on="id_teacher").drop(["id_ratings", "id_student"], axis = 1)
facts_salary_teacher = facts_dropped_teacher.sort_values(by=["salary"], ascending=True).head(5).iloc[:,[2,4,5]]
facts_salary_teacher
```

```python
"""
    You want to know the total number of teachers and students by career
"""
facts_teacher = facts.merge(df_teacher, how="inner", on="id_teacher").drop(["id_ratings", "id_student"], axis = 1)
facts_teacher_race = facts_teacher.merge(df_race, how="inner", on="id_race")
facts_teacher_race.insert(1,'counts',1)
facts_total_teacher_race = facts_teacher_race.groupby(['Profession']).count().reset_index()

facts_student = facts.merge(df_student, how="inner", on="id_student").drop(["id_ratings", "id_teacher"], axis = 1)
facts_student_race = facts_student.merge(df_race, how="inner", on="id_race")
facts_student_race.insert(1,'counts',1)
facts_total_student_race = facts_student_race.groupby(['Profession']).count().reset_index()

facts_total = facts_total_teacher_race.merge(facts_total_student_race, left_on='Profession', right_on='Profession')
facts_total["total"] = facts_total["counts_x"] + facts_total["counts_y"]
facts_total.iloc[:,[0,15]]
```
